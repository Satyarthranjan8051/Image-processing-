```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load your image (replace 'sample image4.jpeg' with your actual image file)
image_path = '/content/sample image4.jpeg'
original_image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)

# Frequency Domain Techniques
def apply_low_pass_filter(image):
    kernel = np.ones((3, 3), np.float32) / 9
    low_pass_filtered = cv2.filter2D(image, -1, kernel)
    return low_pass_filtered

def apply_high_pass_filter(image):
    kernel = np.array([[-1, -1, -1], [-1, 8, -1], [-1, -1, -1]])
    high_pass_filtered = cv2.filter2D(image, -1, kernel)
    return high_pass_filtered

# Spatial Domain Techniques
def apply_gaussian_filter(image):
    gaussian_filtered = cv2.GaussianBlur(image, (5, 5), 0)
    return gaussian_filtered

# Laplacian filter for sharpness detection
def apply_laplacian_filter(image):
    laplacian_result = cv2.Laplacian(image, cv2.CV_64F)
    laplacian_result = np.uint8(np.abs(laplacian_result))
    return laplacian_result

# Sobel edge detection
def apply_sobel_edge_detection(image):
    sobel_x = cv2.Sobel(image, cv2.CV_64F, 1, 0, ksize=3)
    sobel_y = cv2.Sobel(image, cv2.CV_64F, 0, 1, ksize=3)
    sobel_magnitude = np.sqrt(sobel_x**2 + sobel_y**2)
    return sobel_magnitude

# Apply filters
low_pass_result = apply_low_pass_filter(original_image)
high_pass_result = apply_high_pass_filter(original_image)
gaussian_result = apply_gaussian_filter(original_image)
laplacian_result = apply_laplacian_filter(original_image)
sobel_result = apply_sobel_edge_detection(original_image)

# Plot results
plt.figure(figsize=(12, 10))

plt.subplot(2, 3, 1)
plt.imshow(original_image, cmap='gray')
plt.title('Original Image')

plt.subplot(2, 3, 2)
plt.imshow(low_pass_result, cmap='gray')
plt.title('Low Pass Filtered')

plt.subplot(2, 3, 3)
plt.imshow(high_pass_result, cmap='gray')
plt.title('High Pass Filtered')

plt.subplot(2, 3, 4)
plt.imshow(gaussian_result, cmap='gray')
plt.title('Gaussian Filtered')

plt.subplot(2, 3, 5)
plt.imshow(laplacian_result, cmap='gray')
plt.title('Laplacian Filter (Sharpness)')

plt.subplot(2, 3, 6)
plt.imshow(sobel_result, cmap='gray')
plt.title('Sobel Edge Detection')

plt.tight_layout()
plt.show()
```
