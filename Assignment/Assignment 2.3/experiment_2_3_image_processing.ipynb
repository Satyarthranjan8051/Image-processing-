```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
img = cv2.imread('sample image4.jpeg', cv2.IMREAD_GRAYSCALE)
kernel = np.ones((5, 5), np.uint8)
eroded_image = cv2.erode(img, kernel, iterations=1)
```

```python
dilated_image = cv2.dilate(img, kernel, iterations=1)
```

```python
opened_image = cv2.morphologyEx(img, cv2.MORPH_OPEN, kernel)
```

```python
closed_image = cv2.morphologyEx(img, cv2.MORPH_CLOSE, kernel)
```

```python
gradient_image = cv2.morphologyEx(img, cv2.MORPH_GRADIENT, kernel)
```

```python
plt.figure(figsize=(12, 6))

plt.subplot(231), plt.imshow(img, cmap='gray'), plt.title('Original')
plt.subplot(232), plt.imshow(eroded_image, cmap='gray'), plt.title('Erosion')
plt.subplot(233), plt.imshow(dilated_image, cmap='gray'), plt.title('Dilation')
plt.subplot(234), plt.imshow(opened_image, cmap='gray'), plt.title('Opening')
plt.subplot(235), plt.imshow(closed_image, cmap='gray'), plt.title('Closing')
plt.subplot(236), plt.imshow(gradient_image, cmap='gray'), plt.title('Gradient')

plt.tight_layout()
plt.show()
```
